# Visualize pair generation

## Modules and configuration

```python
import sys
import numpy as np
from tensorflow.keras.datasets import mnist
```

```python
sys.path.append('../')
from siamese_networks.build_siamese_pairs import make_pairs
from siamese_networks.viz_helpers import pair_plot
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

## Dataset loading

```python
print("[INFO] loading dataset...")
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

## Pairs dataset for train/test

```python
# build the positive and negative image pairs
print("[INFO] preparing positive and negative pairs...")
(x_pair_train, y_pair_train) = make_pairs(x_train, y_train)
(x_pair_test, y_pair_test) = make_pairs(x_test, y_test)
```

```python
N = 49
sample_idx = np.random.choice(np.arange(0, len(x_pair_train)), size=N)
pairs = x_pair_train[sample_idx]
labels = y_pair_train[sample_idx]
```

```python
pair_plot(pairs, labels, items_per_row=7)
```
